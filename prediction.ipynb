```python
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import TextVectorization

# Configuration

VOCAB_SIZE = 10000
MAX_LEN = 200

# Custom Standardization

def custom_standardization(text):

    text = tf.strings.lower(text)

    text = tf.strings.regex_replace(
        text,
        "<br />",
        " "
    )

    text = tf.strings.regex_replace(
        text,
        "[^a-zA-Z ]",
        ""
    )

    return text

# Load Vocabulary

with open("vocab.pkl", "rb") as f:
    vocab = pickle.load(f)

# Recreate Text Vectorizer

vectorizer = TextVectorization(
    standardize=custom_standardization,
    max_tokens=VOCAB_SIZE,
    output_mode="int",
    output_sequence_length=MAX_LEN
)

vectorizer.set_vocabulary(vocab)

# Load Trained Model

model = load_model("imdb_model.keras")

print(" Model Loaded Successfully!")

# Prediction Function

def predict_sentiment(review):

    # Convert review to Tensor
    review_tensor = tf.constant([review])

    # Vectorize review
    review_vector = vectorizer(review_tensor)

    # Predict
    prediction = model.predict(
        review_vector,
        verbose=0
    )[0][0]

    sentiment = (
        "Positive 😊"
        if prediction >= 0.5
        else "Negative 😞"
    )

    return sentiment, prediction

# Test Examples

reviews = [

    "This movie was fantastic! The acting was brilliant and I loved every minute.",

    "Absolutely amazing movie. One of the best films ever made.",

    "I really enjoyed this movie and would definitely watch it again.",

    "This movie was terrible. I hated every minute of it.",

    "Waste of time. Worst movie ever.",

    "The plot was boring and the acting was awful."

]

for review in reviews:

    sentiment, score = predict_sentiment(review)

    print("=" * 70)
    print("Review:")
    print(review)
    print()
    print(f"Prediction Score : {score:.4f}")
    print(f"Sentiment        : {sentiment}")

print("=" * 70)
```
